```python
def vroce(retezec):
    return rok.with_columns(pl.col('datum').cast(pl.Int8())).join(
    df.filter(pl.col("nazev") == retezec).group_by(pl.col("datum").dt.week()).len(),
    how='left',
    on='datum'
).with_columns(pl.col('len').fill_null(0)).with_columns(pl.lit(retezec).alias('co'))
```

```python
def vroce(retezec):
    # First, create the joined dataframe as you were doing before
    joined_df = rok.with_columns(pl.col('datum').cast(pl.Int8())).join(
        df.filter(pl.col("datum").dt.year() < 2025).filter(pl.col("nazev") == retezec).group_by(pl.col("datum").dt.week()).len(),
        how='left',
        on='datum'
    ).with_columns(pl.col('len').fill_null(0))
    
    # Find the maximum value in the 'len' column
    max_len = joined_df.select(pl.col('len').max()).item()
    
    # If max_len is 0, we'll avoid division by zero
    if max_len == 0:
        return joined_df.with_columns(pl.lit(retezec).alias('co'))
    
    # Otherwise, divide all values by the maximum to get ratio where 1 is the maximum
    return joined_df.with_columns(
        (pl.col('len') / max_len).alias('len'),
        pl.lit(retezec).alias('co')
    )
```

```python
def hledej(retezec):
    return df.filter(pl.col('nazev').str.starts_with(retezec)).group_by('nazev').len().sort(by='len',descending=True).head(20)
```

## Léto / Vánoce

```python
aspon_dva_kanaly = df.group_by('nazev').agg(pl.col('kanal').unique().len()).filter(pl.col('kanal') > 1).select(pl.col('nazev')).to_series().to_list()
aspon_dva_kanaly
```

```python
celorocni = df.group_by(["nazev"]).len().rename({'len':'celorok'})
letni = df.with_columns(pl.col("datum").dt.month().alias("mesic")).filter(pl.col("mesic").is_in([7,8])).group_by(["nazev"]).len().rename({'len':'sezona'})
vanocni = df.with_columns(pl.col("datum").dt.month().alias("mesic")).filter(pl.col("mesic").is_in([1,12])).group_by(["nazev"]).len().rename({'len':'sezona'})
```

```python
celorocni.filter(pl.col('celorok') >= 40).join(letni, how='left', on='nazev').with_columns((pl.col('sezona') / pl.col('celorok')).alias('podil')).sort(by='podil')
```

```python
celorocni.filter(pl.col('celorok') >= 100).join(vanocni, how='left', on='nazev').with_columns((pl.col('sezona') / pl.col('celorok')).alias('podil')).sort(by='podil')
```

```python
import json
with open(os.path.join('data','nazvy_ceskych_filmu.json'), 'r', encoding='utf-8') as dovnitr:
    ceske_filmy = json.loads(dovnitr.read())
```

```python
celorocni.filter(pl.col('celorok') >= 20).join(letni, how='left', on='nazev').with_columns((pl.col('sezona') / pl.col('celorok')).alias('podil')).drop_nulls().sort(by='podil',descending=True).filter(pl.col('nazev').is_in(ceske_filmy)).select(pl.col(['nazev','podil'])).filter(pl.col('podil') > 0.4)
```

```python
celorocni.filter(pl.col('celorok') >= 20).join(vanocni, how='left', on='nazev').with_columns((pl.col('sezona') / pl.col('celorok')).alias('podil')).drop_nulls().sort(by='podil',descending=True).filter(pl.col('nazev').is_in(ceske_filmy)).select(pl.col(['nazev','podil'])).filter(pl.col('podil') > 0.5)
```
