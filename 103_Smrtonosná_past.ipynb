## Je Smrtonosná past vánoční film?

```python
import os
import polars as pl
pl.Config(tbl_rows=100)

df = pl.read_parquet(os.path.join("data","fdb.parquet"))
```

## Co máme v datech

```python
df.group_by("kanal").len().sort(by="len",descending=True)
```

```python
df.group_by(pl.col("datum").dt.year()).len().sort(by="datum")
```

## A kde je tam Smrtonosná past?

```python
df.filter(pl.col("nazev").str.starts_with("Smrtonosná past")).sort(by="datum")
```

```python
df.filter(pl.col("nazev").str.starts_with("Smrtonosná past")).filter(pl.col("datum").dt.month() == 12).sort(by="datum")
```
