```python
import os
import json
```

```python
def konverze(p):
    ## do 6 hodin vrátit o den zpět!

    try:
    
        return [
            {
                'kanal': p.get('@attributes', {}).get('kanal'),
                'datum': p.get('@attributes', {}).get('datum_vysilani'),
                'cas': porad.get('cas'),
                'nadtitul': porad.get('nazvy', {}).get('nadtitul'),
                'nazev': porad.get('nazvy', {}).get('nazev'),
                'nazev_casti': porad.get('nazvy', {}).get('nazev_casti'),
                'nazev_original': porad.get('nazvy', {}).get('original'),
                'zanr': porad.get('zanr'),
                'noticka': porad.get('noticka'),
                'stopaz': porad.get('stopaz'),
                'premiera': porad.get('ikonky', {}).get('premiera'),
                'hvezdicka': porad.get('ikonky', {}).get('hvezdicka'),
                'zive': porad.get('ikonky', {}).get('live'),
                'puvodni_zneni': porad.get('ikonky', {}).get('puvodni_zneni')
            }
            for porad in p.get('porad', [])
        ]

    except:
        
        print(p)
```

```python
odkud = "downloads/ct"
kam = "data_raw/ct"
os.makedirs(kam, exist_ok=True)
hotove = set(os.listdir(kam))
```

```python
pocitadlo = 0
k_prevedeni = [s for s in os.listdir(odkud) if (os.path.getsize(os.path.join(odkud, s)) > 100) and (s.endswith('.json'))]
for f in k_prevedeni:
    if f not in hotove:
        pocitadlo += 1
        if pocitadlo % 1000 == 0:
            print(f"{pocitadlo}/{len(k_prevedeni)}: {f}")
        with open(os.path.join(odkud, f), 'r', encoding="utf-8") as zdrojovy_soubor:
            converted_records = konverze(json.loads(zdrojovy_soubor.read()))
        try:
            with open(os.path.join(kam, f), 'w+', encoding="utf-8") as cilovy_soubor:
                for record in converted_records:
                    # Write each record as a JSON line with a newline character
                    cilovy_soubor.write(json.dumps(record, ensure_ascii=False) + '\n')
        except:
            print(f)
```
